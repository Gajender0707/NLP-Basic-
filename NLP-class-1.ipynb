```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

# Text Preprocessing

```python
nltk.download("punkt")
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
type(sentences)
```

```python
stemmer=PorterStemmer()
```

```python
stemmer.stem("going")
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("histori")
```

```python
import re
```

```python
sentences
```

```python
corpus=[]
for i in sentences:
    review=re.sub("[^a-zA-Z]"," ",i)
    review=review.lower()
    corpus.append(review)
    print(review)
```

```python
corpus
```

```python
from nltk.tokenize import word_tokenize
```

```python
len(corpus)
```

```python
words=[]
for i in range(len(corpus)):
    words.append(word_tokenize(corpus[i]))
```

```python
words
```

```python
len(words)
```

```python
stem_words=[]
for i in words:
    for j in i:
        stem_words.append(stemmer.stem(j))
```

```python
stemmer.stem("serving")
```

```python
stops = set(stopwords.words('english'))
```

```python
stops
```

```python
len(stem_words)
```

```python
final_words=[]
for i in stem_words:
    if i not in stops:
        final_words.append(i)
        
```

```python
s_words=[]
for i in corpus:
    words=word_tokenize(i)
    for j in words:
        if j not in set(stopwords.words('english')):
            word=stemmer.stem(j)
            s_words.append(word)
```

```python
len(s_words)
```

```python
l_words=[]
for i in corpus:
    words=word_tokenize(i)
    for j in words:
        if j not in set(stopwords.words('english')):
            word=lemmatizer.lemmatize(j)
            l_words.append(word)
```
